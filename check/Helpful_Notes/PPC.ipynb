# phases
- Preplanning
  - market survey & forecasting
  - prod. design
  - prod. spec
  - process design
  - plant layout
- Planning : Plan your work with RLS
  - Consists of
    - capacity planning
    - aggregates
    - mats
    - enterprise resources
    - op sequencing
    - tools planning
    - scheduling
    - loading
  - Functions of Planning (RLS)
  - Routing
    - to determine an econ-seq of ops
    - general
      - by building
      - machine shop, assembly etc.
    - detailed 
      - the specific workstation/machine to be used in each op
    - essential
      - production is dependent on routing
  - Loading
    - amt of work 
      - assigned to a machine/worker
    - record of workload
      - of diff. workshops
    - total time
      - (unit op time) * (# of parts to be processed)
    - overload? then
      - OT
      - additional shift intro
      - another shop
      - sub-contracting
  - Scheduling
    - assignment of starting & completion times
      - for the var. ops to be perf.
    - when to be perf.
    - when to be completed
    - preliminary info for a schedule is obtained from
      - planning dpt.
      - date of delivery
      - min. time req for production (past experience)
      - other
        - availability of eq/skill/mats
    - Types of Schedules
      - Plant Schedule
        - reg. a particular plant
      - Master Schedule
        - prod. of a given product
          - in many divisions as desired
        - period : 12 m√©s or more
    - simple to follow & flexible
    - make provision for emergencies
      - rush orders
      - breakdown of machine/power
      - absence of workers
- Control : Work your plan
  - Dispatching, Expediting & Follow-Up, Corrective Action
    - dispatching
    - expediting
    - production follow-up
    - data collection & interpretation
    - progress reporting
    - replanning and modification
